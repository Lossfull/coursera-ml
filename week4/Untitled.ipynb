```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('close_prices.csv')
df
```

```python
pca = PCA(n_components=10)
pca.fit(df.iloc[:,1:])
```

```python
sum(pca.explained_variance_ratio_[0:4])
```

```python
len(pca.components_[0])
```

```python
df2 = pca.transform(df.iloc[:,1:])
```

```python
first_comp = [i[0] for i in df2]
```

```python
df_dija = pd.read_csv('djia_index.csv')
```

```python
df_dija
```

```python
np.corrcoef(first_comp,np.array(df_dija.iloc[:,1]))
```

```python
first_comp.index(max(first_comp))
```
