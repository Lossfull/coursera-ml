### Самое популярное женское имя

```python
df['Name'] = df.Name.str.replace('[^a-zA-Z ]', '')
df['Name'] = df['Name'].str.split()
femnames = df['Name'][df['Sex']=='female']
```

```python
dict = {}
for i in femnames:
    if 'Miss' in i:
        name = i[i.index('Miss')+1]
        if name in dict.keys():
            dict[name] += 1
        else:
            dict[name] = 1
    elif 'Mrs' in i:
        
        name = i[i.index('Mrs')+2] if i.index('Mrs')+2<len(i) else i.index('Mrs')+1 
        if name in dict.keys():
            dict[name] += 1
        else:
            dict[name] = 1
dict
```

```python
values=[]
for i in dict.keys():
    values.append(dict[i])
max(values)
```
