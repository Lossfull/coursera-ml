```python
import numpy as np
import pandas as pd
from sklearn.metrics import roc_auc_score
```

```python
df = pd.read_csv('data-logistic.csv', header=None)
```

```python
y = df.iloc[:,0]
X = df.iloc[:,1:]
```

```python
def log_fun(w1,w2,C):
    return (1/len(df))*sum([np.log10(1+np.exp(row[1][0]*(row[1][1]*w1+row[1][2]*w2))) for row in df.iterrows()]) + C/2*(np.linalg.norm([w1,w2]))

def log_fun_deriv_w1(w1,w2,C=10):
    return (1/len(df))*sum([row[1][0]*row[1][1]*(1-(1/(1+np.exp(row[1][0]*(row[1][1]*w1+row[1][2]*w2))))) for row in df.iterrows()]) - C*w1

def log_fun_deriv_w2(w1,w2,C=10):
    return (1/len(df))*sum([row[1][0]*row[1][2]*(1-(1/(1+np.exp(row[1][0]*(row[1][1]*w1+row[1][2]*w2))))) for row in df.iterrows()]) - C*w2

def euc_distance(x1,x2):
    return np.sqrt(sum([(x1[i]-x2[i])**2 for i in range(len(x1))]))

def grad_descent(w1=0, w2=0, max_iter=2000, eps_targ=1e-5, k=0.1, C=0):
    eps = 1
    for i in range(max_iter):
        w1_last,w2_last = w1,w2
        w1 += k*log_fun_deriv_w1(w1,w2,C=C)
        w2 += k*log_fun_deriv_w2(w1,w2,C=C)
        eps = euc_distance([w1_last,w2_last],[w1,w2])
        if eps<eps_targ:
            break
    return([w1,w2])
        
    
```

```python
w1,w2 = grad_descent()
```

```python
w1,w2
```

```python
prob = [1/(1+np.exp(-w1*df.iloc[i,1]-w2*df.iloc[i,2])) for i in range(len(df))]
roc_auc_score(y,prob)
```
